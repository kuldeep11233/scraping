```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Demographics_of_India"
```

```python
page = requests.get(url)
```

```python
print(page)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[7]
```

```python
print(table)
```

```python
 population = table.find_all('th')
```

```python
population
```

```python
population_groth = [title.text.strip() for title in population]
print(population_groth)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = population_groth)
df
```

```python
columns_data = table.find_all('tr')
```

```python
print(columns_data)
```

```python
for row in columns_data[1:]:
    row_data = row.find_all('td')
    indvisual_data = [data.text.strip() for data in row_data ]
    length = len(df)
    df.loc[length] = indvisual_data
```
